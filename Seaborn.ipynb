```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

from pydataset import data
```

```python
iris= data('iris')
```

```python
iris = sns.load_dataset('iris')
```

```python
#1What does the distribution of petal lengths look like?

sns.distplot(iris.petal_length)
```

```python
#2Is there a correlation between petal length and petal width?
sns.relplot(data=iris, x='petal_length', y='petal_width')
```

```python
#3Would it be reasonable to predict species based on sepal width and sepal length?
sns.jointplot(data=iris, x='petal_length', y='petal_width')
```
